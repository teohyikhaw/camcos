```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from data_personal_example import transaction_data_file, block_data_file,  \
large_pre_gas_prices_file, suite_spot_txn, suite_spot_blx, new_post_eip_txn, \
new_post_eip_blx, new_post_eip_rcpts, post_eip_oct_txn_file, post_eip_oct_blx_file





#post eip later on
post_eip_later_txn=pd.read_csv(new_post_eip_txn)
post_eip_later_blx=pd.read_csv(new_post_eip_blx)
post_eip_later_rpcts=pd.read_csv(new_post_eip_rcpts)


#post eip day 1
post_eip_d1_txn=pd.read_csv(transaction_data_file)
post_eip_d1_blx=pd.read_csv(block_data_file)

#post eip late october
post_eip_oct_txn=pd.read_csv(post_eip_oct_txn_file)
post_eip_oct_blx=pd.read_csv(post_eip_oct_blx_file)


#pre eip gas prices
large_pre_gas_prices=list(pd.read_csv(large_pre_gas_prices_file).gas_price)




#functions
def sumthing(larger_list,smaller_list):
    return sum(smaller_list)/sum(larger_list)

def random_index_generator(my_list,num_of_indexes):
    my_max=len(my_list)-1
    result=np.random.uniform(0,my_max,num_of_indexes)
    return [round(x) for x in result]
```

THE FIRST CHUNK IS FROM THE FIRST DAY OF EIP 1559

```python
#get rid of nulls
post_txns=post_eip_d1_txn[pd.notnull(post_eip_d1_txn.max_priority_fee_per_gas)]
#df[pd.notnull(df.column_of_interest)]

#get max fee
max_fee=list(post_txns.max_fee_per_gas)

#clear max fee outliers
fee_up=np.percentile(max_fee,95)
fee_lo=np.percentile(max_fee,10)
max_fee=[x for x in max_fee if (x < fee_up) & (x > fee_lo)]

#clear pre eip outliers
my_up=np.percentile(large_pre_gas_prices,95)
my_lo=np.percentile(large_pre_gas_prices,10)
my_pre_gas=[x for x in large_pre_gas_prices if (x < my_up) & (x > my_lo)]



#get # of pre eip data, same amount as our post data
my_index=random_index_generator(my_pre_gas,33940)
my_pre_gas=[my_pre_gas[x] for x in my_index]


#plotting
fig = plt.figure()
ax = plt.axes()

my_bins=np.linspace(0, 360000000000,20)

x_ticks=[my_bins[0],round(my_bins[3]),\
round(my_bins[7]),round(my_bins[11]),round(my_bins[15]),my_bins[19]]

plt.hist(max_fee,bins=my_bins, label='Max Fee (Post-Eip)',alpha=0.5,ec='black')
plt.hist(my_pre_gas,bins=my_bins,label='Gas Price (Pre-Eip)',alpha=0.5,ec='black')
plt.xticks(x_ticks)
ax.set_title('Gas Price (Pre-EIP) vs. Max Fee (Post-EIP)')
ax.set_xlabel('Eth Per Unit of Gas (*10^11)')
ax.set_ylabel('Frequency')
plt.legend()


#give stats
plt.savefig('gas_price_v_max_fee.png')
print('Regular: \n')
print('max fee variance=' + str(np.var(max_fee)) + ', pre gas variance=' \
      + str(np.var(my_pre_gas)))


#rescale the data
rescale_ratio=sumthing(my_pre_gas,max_fee)
rescaled_pre_eip_gas_price=[x * rescale_ratio for x in my_pre_gas]


#plotting
fig2 = plt.figure()
ax2 = plt.axes()

my_bins=np.linspace(0, 250000000000,20)

x_ticks=[my_bins[0],round(my_bins[3]),\
round(my_bins[7]),round(my_bins[11]),round(my_bins[15]),my_bins[19]]

plt.hist(max_fee,bins=my_bins, label='Max Fee (Post-Eip)',alpha=0.5,ec='black')
plt.hist(rescaled_pre_eip_gas_price, label='Rescaled Pre EIP Gas Price', \
        alpha=0.5,ec='black',bins=my_bins)
plt.legend()
plt.xticks(x_ticks)
ax2.set_title(' Rescaled Gas Price (Pre-EIP) vs. Max Fee (Post-EIP)')
ax2.set_xlabel('Eth Per Unit of Gas (*10^11)')
ax2.set_ylabel('Frequency')
print('\n \n')
print('rescaled: \n')
print('max fee variance=' + str(np.var(max_fee)) + ', rescaled pre gas variance=' + \
      str(np.var(rescaled_pre_eip_gas_price)))
```

```python
#display(my_txns)

#Get the block number and gas limits and base fee from the dataset
my_block_number=list(post_eip_d1_blx.block_number)
my_gas_limit=list(post_eip_d1_blx.gas_limit)
my_base_fee=list(post_eip_d1_blx.base_fee_per_gas)
#initialize a dictionary to assign gas limits and base fee to the transaction data
gas_limit_tracker={}
base_fee_tracker={}


#makes a dicionary with the key being the block number and the 
#value being the gas limits and base fee, because this will allow us to 
#easily assign a gas limit and base fee to the transaction going forward
for i in range(len(my_block_number)):
    gas_limit_tracker[my_block_number[i]]=my_gas_limit[i]
    base_fee_tracker[my_block_number[i]]=my_base_fee[i]
    
    
##get the block numbers from the transaction data
post_transaction_block_numbers=list(post_txns.block_number)


#initialize a list for the purpose of saving the gas limits that will 
#be assigned to the transaction data and assign the correct information
gas_limits_for_transaction_data=[gas_limit_tracker[x] for x in post_transaction_block_numbers]
base_fee_for_transaction_data=[base_fee_tracker[x] for x in post_transaction_block_numbers]

    
#add the column into the dataframe
#transaction_data['gas_limit']=gas_limits_for_transaction_data
post_txns['base_fee']=base_fee_for_transaction_data
post_txns['base_fee_plus_tip']=post_txns['base_fee']+post_txns['max_priority_fee_per_gas']
display(post_txns)
```

```python
my_tip=list(post_txns.base_fee_plus_tip)



tip_hi=np.percentile(my_tip,95)
tip_lo=np.percentile(my_tip,10)
my_tip=[x for x in my_tip if (x < tip_hi) & (x > tip_lo)]


my_up=np.percentile(large_pre_gas_prices,95)
my_lo=np.percentile(large_pre_gas_prices,10)
my_pre_gas=[x for x in large_pre_gas_prices if (x < my_up) & (x > my_lo)]


#get # of pre eip data, same amount as post data
my_index=random_index_generator(my_pre_gas,33157)
my_pre_gas=[my_pre_gas[x] for x in my_index]


#plotting
fig = plt.figure()
ax = plt.axes()



my_bins=np.linspace(0, 360000000000,20)
x_ticks=[my_bins[0],round(my_bins[3]),\
round(my_bins[7]),round(my_bins[11]),round(my_bins[15]),my_bins[19]]

plt.hist(my_tip,bins=my_bins,alpha=0.5,label='Base Fee + Tip (Post-EIP)',ec='black')
plt.hist(my_pre_gas,bins=my_bins,alpha=0.5,label='Gas Price (Pre-EIP)',ec='black')
ax.set_title('Gas Price (pre-EIP) Vs. Base Fee + Tip (post-EIP)')
ax.set_xlabel('Eth Per Unit of Gas (*10^11)')
ax.set_ylabel('Frequency')
plt.xticks(x_ticks)
plt.legend()

#plt.savefig('gas_price_v_base_fee_plus_tip.png')
print('Regular: \n')
print('base fee + tip variance =' + str(np.var(my_tip)) + ', pre gas variance='  \
      + str(np.var(my_pre_gas)))



rescale_ratio=sumthing(my_pre_gas,my_tip)
rescaled_pre_eip_gas_price=[x * rescale_ratio for x in my_pre_gas]




fig2 = plt.figure()
ax2 = plt.axes()



my_bins=np.linspace(0, 200000000000,20)
x_ticks=[my_bins[0],round(my_bins[3]),\
round(my_bins[7]),round(my_bins[11]),round(my_bins[15]),my_bins[19]]

plt.hist(my_tip,bins=my_bins,alpha=0.5,label='Base Fee + Tip (Post-EIP)',ec='black')
plt.hist(rescaled_pre_eip_gas_price,bins=my_bins,alpha=0.5, \
         label='Gas Price (Pre-EIP)',ec='black')
ax2.set_title('Gas Price (pre-EIP) Vs. Base Fee + Tip (post-EIP)')
ax2.set_xlabel('Eth Per Unit of Gas (*10^11)')
ax2.set_ylabel('Frequency')
plt.xticks(x_ticks)
plt.legend();


print('\n \n')
print('rescaled: \n')
print('base fee + tip variance=' + str(np.var(my_tip)) + ', pre gas variance=' + \
      str(np.var(rescaled_pre_eip_gas_price)))
```
